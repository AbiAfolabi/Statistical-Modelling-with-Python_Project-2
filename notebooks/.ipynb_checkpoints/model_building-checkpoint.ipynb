#### Normal_Bikes is the only column with '0' as a value.

#### Considering the volume of data I have, I will not be dropping any rows instead will be replacing the numerical columns (Free_Bikes, Empty_Slots, Total_Slots, Normal_Bikes, Review_Count  and Rating with their  respective mean value

```python

df_City_Bikes_Merged['Free_Bikes'] = df_City_Bikes_Merged['Free_Bikes'].fillna(value = df_City_Bikes_Merged['Free_Bikes'].mean())

df_City_Bikes_Merged['Empty_Slots'] = df_City_Bikes_Merged['Empty_Slots'].fillna(value = df_City_Bikes_Merged['Empty_Slots'].mean())

df_City_Bikes_Merged['Total_Slots'] = df_City_Bikes_Merged['Total_Slots'].fillna(value = df_City_Bikes_Merged['Total_Slots'].mean())

df_City_Bikes_Merged['Distance'] = df_City_Bikes_Merged['Distance'].fillna(value = df_City_Bikes_Merged['Distance'].mean())

df_City_Bikes_Merged['Normal_Bikes'] = df_City_Bikes_Merged['Normal_Bikes'].fillna(value = df_City_Bikes_Merged['Normal_Bikes'].mean())

df_City_Bikes_Merged['Review_Count'] = df_City_Bikes_Merged['Review_Count'].fillna(value = df_City_Bikes_Merged['Review_Count'].mean())

df_City_Bikes_Merged['Rating'] = df_City_Bikes_Merged['Rating'].fillna(value = df_City_Bikes_Merged['Rating'].mean())
```

```python
df_City_Bikes_Merged
```

### Simple Linear Regression to predict number of Free_Bikes

```python
from sklearn import linear_model
import statsmodels.api as sm
```

```python
x = df_City_Bikes_Merged[['Empty_Slots', 'Total_Slots', 'Normal_Bikes', 'Longitude', 'Latitude', 'Distance', 'Review_Count', 'Rating']]
y = df_City_Bikes_Merged['Free_Bikes']
```

```python
Xreg = sm.add_constant(x) # adding a constant

reg_line = sm.OLS(y, Xreg)

model = reg_line.fit()

result = model.summary()

print(result)
```

Provide model output and an interpretation of the results. 

```python
result
```

R-squared: the proportion of variance in the dependent variable that can be explained by the independent variable (goodness of fit). Ranges from 0 to 1.
Adj. R-squared: This is preferred to the R-Squared because it penalizes the model for  using multiple independent variables.Usually Adjusated R-squared for very good models is expected to be higher than the R-squared. In this example Adjusted R-squared dropped because there are so many variable without explanatory power that are inclded in the model.
    
The only independent variable with a p-value < 0.05 is "Total_Slots


Building a model with only Total_Slots as the independent variable (as shown below), the Adjusted R-squared increased slightly, however, it is still very low. The closer the value is to 1, the better the quality of the model. This further explains that the variables being used to determine the number of Free_Bikes are not good predictors. 

```python
x = df_City_Bikes_Merged[['Total_Slots']]
y = df_City_Bikes_Merged['Free_Bikes']
```

```python
Xreg = sm.add_constant(x) # adding a constant

reg_line = sm.OLS(y, Xreg)

model = reg_line.fit()

result = model.summary()

print(result)
```

# Stretch

How can you turn the regression model into a classification model?

To achieve this we will need a column of our dependent variable to be categorical (not numerical). Thus I will follow the steps below:
- Choose Online column with values True and NaN
- Assign the value '1' to True and '0' to NaN

```python
df_Classification = df_City_Bikes_Merged.copy()

df_Classification
```

```python
df_Classification.info()
```

```python
df_Classification = df_Classification.replace({True: 1,'NaN': 0 })


df_Classification['Online'] = df_Classification['Online'].fillna(0)
```

```python
df_Classification
```

```python
y = df_Classification['Online']
X = df_Classification[['Free_Bikes','Empty_Slots', 'Total_Slots', 'Normal_Bikes', 'Longitude', 'Latitude', 'Distance', 'Review_Count', 'Rating']]
X = sm.add_constant(X) #adds a column of 1's so the model will contain an intercept

model = sm.Logit(y, X) # (need to send in as floats)

results = model.fit() #fit the model (MLE)
print(results.summary())
```
