#### Import necessary Libraries

```python
import pandas as pd

import seaborn as sns

import matplotlib.pyplot as plt
```

```python
#Importing bike stations data exported from city Bikes notebook to csv file data into pandas dataframe

df_Slough_Bike_Stations = pd.read_csv('Slough_Bike_Stations.csv')

df_Slough_Bike_Stations
```

```python
df_Slough_Bike_Stations.rename(columns = {'name':'Station_Name', 'free_bikes': 'Free_Bikes', 'empty_slots': 'Empty_Slots', 'extra.total_slots': 'Total_Slots','extra.normal_bikes':'Normal_Bikes', 'extra.online': 'Online', 'longitude': 'Longitude', 'latitude':'Latitude' }, inplace = True)
```

```python
df_Slough_Bike_Stations
```

```python
FSQ_Bakeries = pd.read_csv('FSQ_Bakeries.csv')

FSQ_Bakeries
```

```python
FSQ_Bakeries.rename(columns ={'name':'Bakery_Name', 'distance': 'Distance', 'geocodes.main.latitude': 'Latitude', 'geocodes.main.longitude':'Longitude', 'location.formatted_address': 'Address' }, inplace = True)
```

```python
FSQ_Bakeries
```

```python
FSQ_Employment_Agency = pd.read_csv('FSQ_Employment_Agency.csv')

FSQ_Employment_Agency
```

```python
FSQ_Employment_Agency.rename(columns = {'name':'Emp_Agency_Name', 'distance':'Distance', 'geocodes.main.latitude': 'Latitude', 'geocodes.main.longitude':'Longitude', 'location.formatted_address':'Address' }, inplace = True)
```

```python
FSQ_Employment_Agency
```

```python
Yelp_bakeries = pd.read_csv('yelp_bakeries.csv')

Yelp_bakeries
```

```python
Yelp_bakeries.rename(columns ={'name':'Bakery_Name', 'review_count': 'Review_Count', 'rating':'Rating', 'coordinates.latitude': 'Latitude', 'coordinates.longitude':'Longitude', 'location.address1': 'Address' }, inplace = True)
```

```python
Yelp_bakeries
```

```python
Yelp_emp_agencies = pd.read_csv('yelp_emp_agencies.csv')
```

```python
Yelp_emp_agencies
```

```python
Yelp_emp_agencies.rename(columns ={'name':'Emp_Agency_Name', 'review_count': 'Review_Count', 'rating':'Rating', 'coordinates.latitude': 'Latitude', 'coordinates.longitude':'Longitude', 'location.address2': 'Address' }, inplace = True)

Yelp_emp_agencies
```

```python
def draw_histograms(dataframe, features, rows, cols):
    fig=plt.figure(figsize=(20,20))
    for i, feature in enumerate(features):
        ax=fig.add_subplot(rows,cols,i+1)
        dataframe[feature].hist(bins=20,ax=ax,facecolor='midnightblue')
        ax.set_title(feature+" Distribution",color='DarkRed')
        ax.set_yscale('log')
    fig.tight_layout()  
    plt.show()
draw_histograms(df_merged,df_merged.columns,8,4)
```

#### Check for linearity between Depend variable (Free_Bikes) & Independent Variables (Empty_Slots, Total_Slots etc)

```python
sns.scatterplot(data=df_merged, 
                x="Empty_Slots", 
                y="Free_Bikes",
                alpha = 0.7)
plt.suptitle("Scatter Plot Comparing Empty_Slots to Free_Bikes")
plt.show()

sns.scatterplot(data=df_merged, 
                x="Total_Slots", 
                y="Free_Bikes",
                alpha = 0.7)
plt.suptitle("Scatter Plot Comparing Total_Slots to Free_Bikes")
plt.show()


sns.scatterplot(data=df_merged, 
                x="Latitude", 
                y="Free_Bikes",
                alpha = 0.7)
plt.suptitle("Scatter Plot Comparing Latitude to Free_Bikes")
plt.show()

sns.scatterplot(data=df_merged, 
                x="Longitude", 
                y="Free_Bikes",
                alpha = 0.7)
plt.suptitle("Scatter Plot Comparing Longitude to Free_Bikes")
plt.show()

```

#### The scatter plots do not really align with the linearity assumption of Linear regression. Therefore linear model might not be able to efficiently explain the data in terms of variability, prediction accuracy. 

#### Check for multicollinearity amongst the Independent Variables (Empty_Slots, Total_Slots, Normal_Bike, Longitude, Latitude etc etc)

```python
sns.scatterplot(data=df_merged, 
                x="Empty_Slots", 
                y="Total_Slots",
                alpha = 0.7)
plt.suptitle("Scatter Plot Comparing Empty_Slots to Free_Bikes")
plt.show()

sns.scatterplot(data=df_merged, 
                x="Total_Slots", 
                y="Longitude",
                alpha = 0.7)
plt.suptitle("Scatter Plot Comparing Total_Slots to Free_Bikes")
plt.show()


sns.scatterplot(data=df_merged, 
                x="Latitude", 
                y="Longitude",
                alpha = 0.7)
plt.suptitle("Scatter Plot Comparing Latitude to Free_Bikes")
plt.show()

sns.scatterplot(data=df_merged, 
                x="Latitude", 
                y="Normal_Bikes",
                alpha = 0.7)
plt.suptitle("Scatter Plot Comparing Longitude to Free_Bikes")
plt.show()
```

## Task 3: Database

Put all your results in an SQLite3 database (remember, SQLite stores its databases as files in your local machine - make sure to create your database in your project's data/ directory!)

```python
import sqlite3 as sqlite

import pandas as pd

from sqlalchemy import create_engine
```

```python
df_merged.info()
```

```python
import sqlite3
from sqlite3 import Error

def create_connection(path):
    connection = None
    try:
        connection = sqlite3.connect(path)
        print("Connection to SQLite DB successful")
    except Error as e:
        print(f"The error '{e}' occurred")

    return connection
```

```python
connection = create_connection("sm_app.sqlite")
```

```python
def execute_query(connection, query):
    cursor = connection.cursor()
    try:
        cursor.execute(query)
        connection.commit()
        print("Query executed successfully")
    except Error as e:
        print(f"The error '{e}' occurred")
```

```python
create_users_table = """
CREATE TABLE IF NOT EXISTS city_bike_merged (
  id INTEGER PRIMARY KEY AUTOINCREMENT,
Station_Name TEXT, 
                Free_Bikes FLOAT,
                Empty_Slots FLOAT,
                Total_Slots FLOAT,
                Normal_Bikes FLOAT,
                Online BOOLEAN,
                Longitude FLOAT,
                Latitude FLOAT,
                Bakery_Name TEXT,
                Distance FLOAT,
                Address TEXT,
                Emp_Agency_Name TEXT,
                Review_Count FLOAT,
                Rating FLOAT
);
"""
```

```python
execute_query(connection, create_users_table) 
```

#### QC to Check data can be retrieved from database

```python
def execute_read_query(connection, query):
    cursor = connection.cursor()
    result = None
    try:
        cursor.execute(query)
        result = cursor.fetchall()
        return result
    except Error as e:
        print(f"The error '{e}' occurred")
```

```python
select_data = "SELECT * from city_bike_merged"
data = execute_read_query(conn, select_data)

for data in data:
    print(data)
```
